```python
df = df.drop('Ticket', axis=1)
df = df.drop('Cabin', axis=1)
df.head()
```

```python
df['Age'] = (1912 - df['Age']) 
df.rename(columns={'Age': 'Date_birth'}, inplace=True)
df.head(5)
```
